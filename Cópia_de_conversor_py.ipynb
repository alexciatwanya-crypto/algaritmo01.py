```python
# Instala a biblioteca PyPDF2 se ainda não estiver instalada
!pip install PyPDF2

from PyPDF2 import PdfReader
import os

# Nome do arquivo PDF de entrada
nome_do_arquivo_pdf = "meuarquivopdf.pdf"

# Nome do arquivo TXT de saída
nome_do_arquivo_txt = "texto_convertido.txt"

# Verifica se o arquivo PDF existe
if not os.path.exists(nome_do_arquivo_pdf):
    print(f"Erro: O arquivo '{nome_do_arquivo_pdf}' não foi encontrado. Por favor, certifique-se de que ele está no diretório correto.")
else:
    try:
        # Abre o arquivo PDF
        reader = PdfReader(nome_do_arquivo_pdf)

        texto = ""

        # Percorre todas as páginas do PDF e extrai o texto
        for pagina in reader.pages:
            texto += pagina.extract_text() if pagina.extract_text() else ""

        # Salva o texto extraído em um arquivo TXT
        with open(nome_do_arquivo_txt, "w", encoding="utf-8") as arquivo:
            arquivo.write(texto)

        print(f"O PDF '{nome_do_arquivo_pdf}' foi convertido com sucesso para '{nome_do_arquivo_txt}'.")
    except Exception as e:
        print(f"Ocorreu um erro durante a conversão do PDF: {e}")
```
